```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import random
import numpy as np
```

```python
theRes = []
numTrials = int(1.0e5)

for i in range(numTrials):
    cnt = 0
    while 1:
        thisToss = np.random.randint(0,2)
        cnt += 1
        if thisToss == 1: # P(thisToss = 1) = 1/2 (i.e. 0.5)
            theRes.append(cnt)
            break

p = np.mean(theRes)
print 'Mean: ', 1.0/p, '\nStd Dev: ', np.sqrt(abs(1.0-p)/(p**2))

plt.figure()
plt.hist(theRes, bins = 10)
plt.show()
```

```python
theRes = []
numTrials = int(1.0e5)

for i in range(numTrials):
    cnt = 0
    while 1:
        thisToss = np.random.randint(0,2,(1,2))
        cnt += 1
        if np.sum(thisToss) == 2: # P(sum = 2) = 1/4 (i.e. 0.25)
            theRes.append(cnt)
            break

p = np.mean(theRes)
print 'Mean: ', 1.0/p, '\nStd Dev: ', np.sqrt(abs(1.0-p)/(p**2))

plt.figure()
plt.hist(theRes, bins = 10)
plt.show()
```

```python
theRes = []
numTrials = int(1.0e5)

for i in range(numTrials):
    cnt = 0
    while 1:
        thisToss = np.random.randint(0,2,(1,3))
        cnt += 1
        if np.sum(thisToss) == 0: # P(sum = 0) = 1/8 (i.e. 0.125)
            theRes.append(cnt)
            break

p = np.mean(theRes)
print 'Mean: ', 1.0/p, '\nStd Dev: ', np.sqrt(abs(1.0-p)/(p**2))

plt.figure()
plt.hist(theRes, bins = 10)
plt.show()
```
