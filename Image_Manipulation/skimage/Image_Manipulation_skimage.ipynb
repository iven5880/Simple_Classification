```python
%reset -f

from skimage import io
from skimage.io import imread, imsave, show

import numpy as np

import os
os.system('rm guy_*'); # remove all existing images
```

```python
im = imread('guy.jpg')
print 'The dimensions of the image:', im.shape
```

```python
# block out the center of the image with a single (randomly chosen) color

im = imread('guy.jpg')

xl = 1*im.shape[0]/10
xh = 9*im.shape[0]/10

yl = 1*im.shape[1]/10
yh = 9*im.shape[1]/10

r1 = np.random.randint(256)
r2 = np.random.randint(256)
r3 = np.random.randint(256)

im[xl:xh,yl:yh] = (r1,r2,r3)

imsave('guy_New_1.jpg', im)

io.imshow('guy_New_1.jpg')
io.show()
```

```python
# convert each pixel to the average of original RGB values

im = imread('guy.jpg')

xp = im.shape[0]
yp = im.shape[1]

for i in range(xp):
    for j in range(yp):
        tc = sum(im[i,j])/3
        im[i,j] = [tc,tc,tc]

imsave('guy_New_2.jpg', im)

io.imshow('guy_New_2.jpg')
io.show()
```

```python
# create an image made up of randomly selected RGB values

im = imread('guy.jpg')

xp = im.shape[0]
yp = im.shape[1]

for i in range(xp):
    for j in range(yp):
        r1 = np.random.randint(256)
        r2 = np.random.randint(256)
        r3 = np.random.randint(256)
        im[i,j] = [r1,r2,r3]

imsave('guy_New_3.jpg', im)

io.imshow('guy_New_3.jpg')
io.show()
```
