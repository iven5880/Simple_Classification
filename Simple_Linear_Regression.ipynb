# Simple Linear Regression

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# read in and pre-process the dataset

df = pd.read_csv('LinRegData.txt')
df.insert(0, 'Bias Term', 1) # insert the bias term
X = df.drop('Price', axis = 1)
y = df['Price']

X = X.values
X = X.astype(float)

y = y.values
y.astype(float)
y.shape = np.shape(y)[0]
```

```python
# visualize the dataset
# NOTE: this is only a plot of square-footage vs price (the other features are not plotted)

plt.figure(figsize=(12,8))
plt.plot(X[:,1],y, 'o')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('Square-Footage vs Price (Just Data)', fontsize=25)
plt.show()
```

```python
model = LinearRegression(fit_intercept=False, normalize=True)
model.fit(X,y)
yp = model.predict(X)

print 'R^2 score: ', model.score(X,y)
```

```python
# visualize the prediction
# NOTE: this is only a plot of square-footage vs price (the other features are not plotted)

yt = X[:,1] * model.coef_[1]

plt.figure(figsize=(12,8))
plt.plot(X[:,1],y, 'o')
plt.plot(X[:,1],yt, 'r')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('Square-Footage vs Price (Data and Prediction)', fontsize=25)
plt.show()
```
