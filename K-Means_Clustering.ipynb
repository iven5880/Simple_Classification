## Simple K-Means Clustering

This notebook illustrates a simple implementaiton of the **K-Means** clustering algorithm.
A set of **N** points on a plane are generated as an **Nx2** array of uniformly distributed real numbers. These points are then divided into a pre-specified number of clusters.



```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
```

```python
# Generate and visualize the points

thePoints = 1.0 + 9.0*np.random.rand(2000,2)

plt.figure(figsize=(12,8))
plt.plot(thePoints[:,0], thePoints[:,1], 'o')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('Before Clustering', fontsize=25)
plt.show()
```

```python
# Distribute the points into clusters

numClusters = 4

theModel = KMeans(n_clusters=numClusters, init='k-means++', precompute_distances='auto', verbose=0)
theClusters = theModel.fit_predict(thePoints)

plt.figure(figsize=(12,8))
for i in range(numClusters):
    plt.plot(thePoints[theClusters==i,0], thePoints[theClusters==i,1], 'o')


plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('After Clustering', fontsize=25)
plt.legend(['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4'], loc=1, fontsize=20, markerscale=1.5, frameon=True, bbox_to_anchor=(1.3, 1))
plt.show()
```
